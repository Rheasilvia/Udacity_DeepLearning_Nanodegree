## 线性回归注意事项
线性回归隐含一系列的前提假设，并非适合所有情形，因此应当注意以下两个问题。

### 最适用于线性数据
线性回归会根据训练数据生成直线模型，如果训练数据包含非线性关系，你需要选择：**调整数据（新型数据转换）、增加特征数量或改用其他模型。**

### 容易受到异常值影响
线性回归的目标是求取对训练数据而言的“最优拟合”直线，如果数据集中存在不符合总体规律的异常值，最终结果将会存在不小偏差。

>要特别小心异常值

## 简单线性和高次多项式选择

区别：
1. 简单线性误差大，多项式误差小

通过数字函数找到增加误差的方法
1. 简单线性公式：$3x_1+4x_2 + 5 = 0$
2. 多项式$2x_1^3 -2x_1^2x_2-4x_2^3+3x1^2+6x_1x_2+4x_2^2+5=0$

简单模型更具有通用性。

## L1正则化
这种简单的方法叫做L1正则化

**得到方程系数并把绝对值加起来**
$$Error = |2|+|-2|+|-4|+|3|+|6|+|4| =21$$
远大于简单公式误差7，因此复杂模型得到较大的误差。

## L2正则化

**把系数的平方相加**
$$Error = 2^2+(-2)^2+(-4)^2+(3)^2+(6)^2+(4)^2 =85$$
简单模型误差为9+16=25，因此还是复杂模型得到较大的误差。


因此需要加入一个$\lambda$调和误差。
$$ERROR = \lambda L1|L2$$

**如果$\lambda$较大，那么选择L1正则，如果$\lambda$较小，那么选择L2正则。**


### 注意事项
1. L1计算复杂，尽管没有平方运算，但是绝对值很难区分，但是L2中平方是比较不错的导数，计算起来更容易。
2. 只有数据稀疏时候L1正则会比L2正则更快。
3. 如果很多列包含0，那么L1正则化更好些。
4. 如果数据均匀分布在各列，那么L2正则化更好些。

**L1正则化最大的好处在于提供特征选择，比如有1000个数据，实际只有10个相关，其他都是噪音，L1可以把不相关的数据转变为0.**

#### 稀疏数据
比如你有1000次数据运算，实际只有10个相关项。那么L1比较合适。
