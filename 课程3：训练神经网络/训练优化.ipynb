# 解决过拟合的方法

## 正则化
### 简单规则

L1正则化：
- 希望得到稀疏向量
- 想要降低权重值，最终得到较小的数

L2正则化：
- 不支持稀疏化

## dropout
有些神经网络权重很大，因此一直被训练，而有些神经元一直没被训练，因此可以用dropout关闭一部分，让网络其他的部分也接受训练，从而提高泛化性解决过拟合问题

# 如何解决梯度下降局部最小值问题？梯度消失
## 为什么出现这种情况？
因为有些单元权重很小，那么所计算的梯度也会非常小，导致每次梯度下降的时候变化很小很小。就像我们迈出很小的步子，永远无法从珠峰上下来。

## 1.替换激活函数
最佳方式：
- 替换激活函数
    - 双曲线正切函数$tanh(x)=\frac{e^x-e^{-x}}{e^x+e^{-x}}
    - 修正线性单元relu
    
> - 如果最后一个激活函数是sigmoid函数，那么输出0，1的概率，是分类的问题。
> - 如果最后一个激活函数是relu函数，那么会是一个回归问题.

## 2.批次和随机梯度下降

每一步叫做epochs,通常指的步长实际上就是**epochs的次数**。

> - 小批次导致每次计算精度下降，小批量执行多次步长，相当于全部数据执行一次步长。
> - **但是在现实中，采取大量稍微不准确的步长比采取一次很准确的步长要好很多**。

## 3.学习率的设置问题
1. 学习率过大
学习率太大，那么采取的就是很大的步长。一开始速度会很快，但是会错过**最低值**，使得模型很混乱
2. 学习率过小
导致模型更新速度很慢，虽然能到达最小值，但是会很慢很慢.

> 这里有条好的经验：
> 如果模型不可行，那么就**降低学习率**。



** 最好的学习率是开始学习率较大，当模型越来越接近解决方案时，会降低学习率。**


## 4.随机梯度下降
原理很简单，随机从不同的低点开始，从所有的这些低点进行梯度下降，**增大抵达全局最低点概率**
